# Visualisasi Diferensial dan Integral

Notebook ini menunjukkan visualisasi sederhana dari konsep turunan (diferensial) dan integral dari suatu fungsi matematika.

Kita menggunakan fungsi info:

    f(x) = sin(x)

dan menampilkan:

- Grafik fungsi asli
- Grafik turunannya
- Area di bawah kurva (integral) antara dua titik


```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sin, diff, integrate, lambdify, tan
from sympy.abc import x

# Fungsi f(x)
f = sin(x)
f_prime = diff(f, x)
f_integral = integrate(f, x)

# Ubah jadi fungsi numerik
f_lamb = lambdify(x, f, modules=['numpy'])
f_prime_lamb = lambdify(x, f_prime, modules=['numpy'])

# Rentang x untuk plot
x_vals = np.linspace(0, 2 * np.pi, 500)
y_vals = f_lamb(x_vals)
y_prime_vals = f_prime_lamb(x_vals)
```

```python
# Visualisasi fungsi dan turunannya
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = sin(x)', color='blue')
plt.plot(x_vals, y_prime_vals, label="f'(x) = cos(x)", color='red', linestyle='--')
plt.axhline(0, color='black', linewidth=0.5)
plt.title('Fungsi dan Turunannya')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Area di bawah kurva dari x=0 sampai x=pi
a, b = 0, np.pi
x_area = np.linspace(a, b, 300)
y_area = f_lamb(x_area)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = sin(x)', color='blue')
plt.fill_between(x_area, y_area, color='skyblue', alpha=0.4, label='Area di bawah kurva')
plt.axhline(0, color='black', linewidth=0.5)
plt.title('Integral dari f(x) antara x=0 dan x=π')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Nilai integral simbolik
from sympy import pi
integral_val = integrate(f, (x, 0, pi))
print(f"∫ sin(x) dx dari 0 sampai π = {integral_val}")
```
